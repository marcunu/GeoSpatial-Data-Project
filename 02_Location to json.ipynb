```python
df_schools = pd.read_json("json/schools.json")
df_dog = pd.read_json("json/hairdresser.json")
df_star = pd.read_json("json/starbucks.json")
df_vegan = pd.read_json("json/restaurantes_veganos.json")
df_airport = pd.read_json("json/airport.json")
df_basket = pd.read_json("json/basketball.json")
df_pubs = pd.read_json("json/pub.json")
```

```python
df_schools["type"] = "school"
df_dog["type"] = "dog_hairdresser"
df_star["type"] = "starbucks"
df_vegan["type"] = "vegan_restaurant"
df_airport["type"] = "airport"
df_basket["type"] = "basket"
df_pubs["type"] = "pub"
```

```python
data = pd.concat([df_schools, df_dog, df_star, df_vegan, df_airport, df_basket, df_pubs])
```

```python
ubi = data.reset_index(drop=True)
```

```python
def lat(coord):
    return coord.get("coordinates")[0]

def lon(coord):
    return coord.get("coordinates")[1]
```

```python
ubi["latitud"] = ubi["location"].apply(lt.lat)
ubi["longitud"] = ubi["location"].apply(lt.lon)
ubi
```

```python
ubi.to_csv("./data/nearby.csv")
```
